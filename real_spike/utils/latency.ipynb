## Total Latency

```python
gen_y = [a/1e9 for a  in list(gen_df["latency"])][1:]
```

```python
proc_y = [a/1e9 for a  in list(processor_df["latency"])]
```

```python
viz_y =  [a/1e9 for a  in list(viz_df["latency"])]
```

```python
total_latency = [x + y + z for x, y, z in zip(gen_y, proc_y, viz_y)]
```

```python
plt.hist(total_latency)
plt.ylabel("count")
plt.xlabel("time (s)")
plt.title("Total Latency")
```
