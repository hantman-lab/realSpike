## Online fetching

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_hdf("/home/clewis/repos/realSpike/real_spike/acquisition/latency_df.h5")
```

```python
df
```

## Restrict to 30kHz sampling

```python
sub_df = df[df["downsample_factor"] == 1]
```

```python
sub_df
```

## Look at channels

```python
s = sub_df[sub_df["num_channels"] == 120]
fig, ax = plt.subplots(figsize=(10, 8))
bars = ax.bar(s["num_samples"] / 30, round(s["avg_latency"], 3))

# Add labels to the bars
ax.bar_label(bars)

plt.ylabel("Time (ms)")
plt.xlabel("Fetch Size (ms)")
plt.title("Average Latency (1000 fetches)")
ax = plt.gca()
plt.text(0.05, 0.95, f'120 Channels', 
         transform=ax.transAxes, 
         ha='left', va='top',
         fontsize=10, fontweight='bold')
```
